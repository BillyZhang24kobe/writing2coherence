```python
import os
from ast import literal_eval
import pandas as pd
```

```python
dc = pd.read_csv('data/train_per_sent.csv')
dc.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
dc.head()
```

```python
def get_context(sents: list) -> list:
    context = [sents[0]]
    for sentence in sents[1:-1]:
        context.append(context[-1] + " " + sentence)
    context.insert(0, "")
    return context
```

```python
dc['sents'] = dc['sents'].apply(literal_eval)
dc['context'] = dc['sents'].apply(get_context)
dc['coherence_per_sent'] = dc['coherence_per_sent'].apply(literal_eval)
dc['cohesion_per_sent'] = dc['cohesion_per_sent'].apply(literal_eval)
dc['consistency_per_sent'] = dc['consistency_per_sent'].apply(literal_eval)
dc['relevance_per_sent'] = dc['relevance_per_sent'].apply(literal_eval)
```

```python
dc['coherence_per_sent'] = dc['coherence_per_sent'].apply(lambda item: list(item.values()))
dc['cohesion_per_sent'] = dc['cohesion_per_sent'].apply(lambda item: list(item.values()))
dc['consistency_per_sent'] = dc['consistency_per_sent'].apply(lambda item: list(item.values()))
dc['relevance_per_sent'] = dc['relevance_per_sent'].apply(lambda item: list(item.values()))
```

```python
dc = dc.explode(['context', 'sents', 'coherence_per_sent', 'cohesion_per_sent', 'consistency_per_sent', 'relevance_per_sent'])
```

```python
dc = dc.rename({'title': 'topic', 'sents': 'sentence'}, axis=1)
dc[['topic', 'context', 'sentence', 'coherence_per_sent','cohesion_per_sent','consistency_per_sent','relevance_per_sent']].to_csv('cohesentia_train.csv')
```

```python
dc.head()
```

```python
dc = dc.rename({'topic': 'title'}, axis=1)
dc['sents'] = dc.apply(lambda row: [row['context'], row['sentence']], axis=1)
dc['text'] = dc.apply(lambda row: row['context'] + " " + row['sentence'], axis=1)
dc['coherence_per_sent'] = dc['coherence_per_sent'].map(lambda item: {'sent0': False, 'sent1': item})
dc['cohesion_per_sent'] = dc['cohesion_per_sent'].map(lambda item: {'sent0': False, 'sent1': item})
dc['consistency_per_sent'] = dc['consistency_per_sent'].map(lambda item: {'sent0': False, 'sent1': item})
dc['relevance_per_sent'] = dc['relevance_per_sent'].map(lambda item: {'sent0': False, 'sent1': item})
```

```python
dt = pd.read_csv('data_relevance/train_per_sent.csv', index_col=0)
dt.head()
```

```python
combined = pd.concat([dt, dc[['title', 'text', 'sents', 'coherence_per_sent', 'cohesion_per_sent', 'consistency_per_sent', 'relevance_per_sent']]], ignore_index=True)
assert len(combined) == len(dc) + len(dt)

os.makedirs('data_relevance_combined', exist_ok=True)
combined.to_csv('data_relevance_combined/train_per_sent.csv')
```
